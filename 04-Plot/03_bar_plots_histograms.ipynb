```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basic Plot
nums = np.random.uniform(size=10)
print(nums)
print(np.arange(10))
plt.bar(np.arange(10),height=nums)
```

```python
# Horizontal Plot
plt.barh(np.arange(10), nums) # No olvidar de cambiar height por width
```

```python
# Horizontal Plot
plt.bar(np.arange(10), nums) # width
```

```python
# Colores y edge color
# No olvidar de cambiar height por width
plt.bar(np.arange(10),
         nums, 
         color='limegreen', 
         edgecolor='maroon') # los edgecolor y color pueden valer none
```

```python
# Alineacion
plt.bar(np.arange(10), nums,color='y', 
         edgecolor='g', align='edge')  # edge
```

```python
# hatch & fill
plt.bar(np.arange(10), nums, hatch='o')
```

```python
plt.bar(np.arange(10), nums, hatch='x', color='none', edgecolor='g')
```

```python
# Width de la barra 1 > 0 
plt.bar(np.arange(10), nums, width=0.5, color='r', edgecolor='y') # 1 0.1 2 
```

```python
# Width de la barra 1 > 0 
plt.bar(np.arange(10), nums, 
        bottom=1*nums,
       color='r',
       edgecolor='maroon') # Tamano del espacio en el bottom
plt.bar(np.arange(10), 0.1*nums, 
        bottom=1*nums,
       color='y',
       edgecolor='g') # Tamano del espacio en el bottom
```

```python
rands = np.random.normal(size=int(1e6))
plt.hist(rands)
```

```python
print(plt.hist(rands))
plt.clf()
```

```python
plt.hist(rands, range=(-4,-2))  # 4 3 2 1 0.5
```

```python
plt.hist(rands, bins=[0,1,2,3,4,5,6,7],histtype='barstacked')  # [0,1,2,3,4,5,6,7]
```

```python
plt.hist(rands, cumulative=True) 
```

```python
plt.hist((rands, 0.5*rands, 0.7*rands, 0.9*rands), 
         bins='auto', 
         histtype='bar') # bar barstacked step stepfilled
```

```python
a = plt.bar(np.arange(10), nums, 
        bottom=1*nums,
       color='r',
       edgecolor='maroon') # Tamano del espacio en el bottom
b = plt.bar(np.arange(10), 0.7*nums, 
       color='y',
       edgecolor='g')
c = plt.bar(np.arange(10), 0.5*nums, 
       color='b',
       edgecolor='maroon',
       label=r"$y = \cos(x)$")

plt.legend((a[0], b[0], c[0]), ('Men', 'Women','child'))

plt.xlabel("$Fechas$")
plt.ylabel(r"$Numeros$")
```
