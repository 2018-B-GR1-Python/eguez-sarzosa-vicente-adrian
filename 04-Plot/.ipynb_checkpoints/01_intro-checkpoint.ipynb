```python
import numpy as np
import matplotlib.pyplot as plt

# Configuracion general de los plots revisar documentacion
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basico
xvals = np.arange(0,10,0.1)
plt.plot(xvals)
```

```python
# Ejes
plt.plot(xvals, np.sin(xvals))
```

```python
# plt.plot(xvals, np.sin(xvals))
# plt.plot(xvals, np.cos(xvals))
plt.plot(xvals, np.tan(xvals))
plt.xlabel("$Valores$")
plt.ylabel(r"$Hols$")
```

```python
# Labels en el grafico Color
plt.plot(xvals, 
         np.sin(xvals), 
         label=r"$y = \sin(x)$")
plt.plot(xvals, 
         np.cos(xvals), 
         label=r"$y = \cos(x)$")
plt.plot(xvals, 
         np.tan(xvals), 
         label=r"$y = \tan(x)$")

plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# Estilo de la linea
# linestyle
plt.plot(xvals, 
         np.sin(xvals), 
         label=r"$y = \sin(x)$", 
         color='#d03806')  # k = black b = blue r = red etc...
# Marcadores -> https://matplotlib.org/api/markers_api.html
plt.plot(xvals, np.cos(xvals), 
         label=r"$y = \cos(x)$", 
         marker='x')

plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend(loc='upper left')
```

```python
# Con el 3er argumento
plt.plot(xvals, np.sin(xvals), 'g--', label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), 'bx', label=r"$y = \cos(x)$")
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# Con el 3er argumento
plt.plot(xvals, np.sin(xvals), 'g--', label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), 'bx', label=r"$y = \cos(x)$")
plt.ylim(0,2)
plt.xlim(0,1)
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend()
```
