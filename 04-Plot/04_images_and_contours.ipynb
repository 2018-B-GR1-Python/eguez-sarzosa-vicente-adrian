```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
```

```python
plt.imshow(rands2d)
plt.colorbar()
```

```python
plt.imshow(rands2d, extent=[0,100,-100, 0])
```

```python
plt.imshow(rands2d, vmax=0.3, vmin= 0) 
plt.colorbar()
```

```python
plt.imshow(rands2d, extent=[0,100,-100, 0], vmax=0.1, vmin= -0.05) 
plt.colorbar()
```

```python
plt.imshow(rands2d, aspect=2) # 0.... 1 horizontal, vertical
```

```python
srands2d = gaussian_filter(rands2d, sigma=100)
plt.imshow(srands2d)
plt.colorbar()
```

```python
plt.contour(srands2d)
plt.colorbar()
```

```python
plt.contour(srands2d, levels=[0,0.003, 0.0045])
```

```python
plt.contour(srands2d, 30)
```

```python
plt.contour(srands2d, 30, colors=['r','b'])
```

```python
plt.contourf(srands2d, 10)
```

```python
plt.contourf(srands2d, 15, hatches=['*','-','.'])
```

```python
plt.contour(srands2d, linestyles=[':','--','-'])
```

```python
contours = plt.contour(srands2d, 13)
plt.clabel(contours)
```
