```python
archivo_books='/Users/usrdel/Documents/GitHub/eguez-sarzosa-vicente-adrian-python/06-Recommendation-Algorithm/data/BX-Books.csv'
```

```python
books=pd.read_csv(
    archivo_books,
    sep=";",
    header=0,
    error_bad_lines=False,
    usecols=[0,1,2],
    index_col=0,
    names=['isbn',"title","author"],
    encoding = "ISO-8859-1")
```

```python
books.head()
```

```python
def book_titulo_autor(isbn):
    titulo = books.at[isbn,"title"]
    autor = books.at[isbn,"author"]
    return titulo, autor
```

```python
book_titulo_autor("0393045218")
```

```python
def n_libros_favoritos_por_usuario(usuario,N):
    ratings_usuario = data[data["user"]==usuario]  # Filtering by user  (relevant data)
    ratings_usuario_ordenado = pd.DataFrame.sort_values(
        ratings_usuario,
        ['rating'],
        ascending=[0])[:N]  # Sort by rating in desc order pick N
    
    ratings_usuario_ordenado["title"] = ratings_usuario_ordenado["isbn"].apply(book_titulo_autor)  # Apply bookMeta to ISBN column
    
    return ratings_usuario_ordenado
```

```python
data = data[data["isbn"].isin(books.index)]  # eliminando inconsistencias
```

```python
ratings_por_usuario = data.user.value_counts()
ratings_por_usuario.head(10)
```

```python
data = data[data["isbn"].isin(ratings_por_libro[ratings_por_libro>10].index)] 
```

```python
data = data[data["user"].isin(ratings_por_usuario[ratings_por_usuario>10].index)]
```

```python
matriz_rating_por_usuario =pd.pivot_table(data, 
                                          values='rating',
                                          index=['user'], 
                                          columns=['isbn'])  # Rating Matrix (pivot)
```

```python
import numpy as np
```

```python
def ditancia_entre_dos_usuarios(usuario_1,usuario_2):
        try:
            ratings_usuario_1 = matriz_rating_por_usuario.transpose()[usuario_1]
            ratings_usuario_2 = matriz_rating_por_usuario.transpose()[usuario_2]
            distancia = hamming(ratings_usuario_1,ratings_usuario_2)
        except: 
            distancia = np.NaN
        return distancia 
```

```python
ditancia_entre_dos_usuarios(usuario_1,usuario_2)
```

```python
usuario_a_recomendar = 204622
todos_usuarios = pd.DataFrame(matriz_rating_por_usuario.index)
todos_usuarios = todos_usuarios[todos_usuarios.user!=usuario_a_recomendar]  # Remove the actual user to recommend
todos_usuarios.head()
```

```python
todos_usuarios["distance"] = todos_usuarios["user"].apply(
    lambda usuario_file_actual: ditancia_entre_dos_usuarios(usuario_a_recomendar,usuario_file_actual))
```

```python
todos_usuarios
```

```python
K = 10
K_valores_mas_cercanos = todos_usuarios.sort_values(
    ["distance"],
    ascending=True)["user"][:K]
```

```python
K_valores_mas_cercanos
```

```python
def nearest_neighbors(usuario,K=10):
    todos_usuario = pd.DataFrame(matriz_rating_por_usuario.index)
    todos_usuario = todos_usuario[todos_usuario.user!=usuario]
    todos_usuario["distance"] = todos_usuario["user"].apply(
        lambda usuario_fila_actual: ditancia_entre_dos_usuarios(usuario,usuario_fila_actual))
    K_usuarios_mas_cercanos = todos_usuario.sort_values(
        ["distance"],
        ascending=True)["user"][:K]
    return K_usuarios_mas_cercanos
```

```python
nearest_neighbors(usuario_a_recomendar,5)
```

```python
K_usuarios_mas_cercanos = nearest_neighbors(usuario_a_recomendar)
K_usuarios_mas_cercanos

N_ratings= matriz_rating_por_usuario[matriz_rating_por_usuario.index.isin(K_usuarios_mas_cercanos)]  # Ratings of the Nearest Neighbors for all books
N_ratings
```

```python
avg_rating = N_ratings.apply(np.nanmean).dropna()  # Average rating per isbn, nanmean is applied for each column (isbn) dropna() drop books which do not have rating
avg_rating.head()
```

```python
libros_ya_leidos = matriz_rating_por_usuario.transpose()[usuario_a_recomendar].dropna().index  # Get ratings of the active user, drop books without rating
libros_ya_leidos
```

```python
avg_rating = avg_rating[~avg_rating.index.isin(libros_ya_leidos)]
```

```python
N=3
topNISBNs = avg_rating.sort_values(ascending=False).index[:N]
```

```python
pd.Series(topNISBNs).apply(book_titulo_autor)  # Tell us title and author
```

```python
def topN(usuario,N=3):
    K_usuarios_mas_cercanos = nearest_neighbors(usuario)
    N_ratings = matriz_rating_por_usuario[matriz_rating_por_usuario.index.isin(K_usuarios_mas_cercanos)]
    avg_rating = N_ratings.apply(np.nanmean).dropna()
    libros_ya_leidos = matriz_rating_por_usuario.transpose()[usuario].dropna().index
    avg_rating = avg_rating[~avg_rating.index.isin(libros_ya_leidos)]
    top_N_libros = avg_rating.sort_values(ascending=False).index[:N]
    return pd.Series(top_N_libros).apply(book_titulo_autor)
```

```python
topN(usuario_a_recomendar,6)
```
